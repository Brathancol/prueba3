```python
  from itertools import cycle
  def digito_verificador(rut):
    rut = str(rut)
    reversed_digits = map(int, reversed(str(rut)))
    factors = cycle(range(2, 8))
    s = sum(d * f for d, f in zip(reversed_digits, factors))
    if (-s) % 11 == 10:
        return "K"
    return (-s) % 11

  def validar_rut(rut):
    rut = str(rut) + "-" + str(digito_verificador(rut))
    return rut

  opciones = ["Comprar departamento", "Mostrar departamentos disponibles", "Ver listado de compradores", "Mostrar ganacias totales", "Salir"]
  cliente_comprador = {}
  opcion = 0
  total = 0
  A = 3800
  B = 3000
  C = 2800
  D = 3500
  cantidad = 0
  total_todo = total * cantidad
  print("---Bienvenidos a Casa Feliz---")

  while opcion != 5:
    print("Porfavor ingresar la opcion que desea realizar: \n")
    for i in range(len(opciones)):
        print(f"{i + 1}) {opciones[i]}")
    opcion = int(input("Seleccione la opcion que desea realizar: "))

    if opcion == 1:
      print(" ____________")
      print("|piso| tipo  |")
      print("|    |A|B|C|D|")
      print("| 10 | | | | |")
      print("|____________|")
      print("|  9 | | | | |")
      print("|____________|")
      print("|  8 | | | | |")
      print("|____________|    *   Tipo A, 3.800 UF")
      print("|  7 | | | | |    *   Tipo B, 3.000 UF")
      print("|____________|    *   Tipo C, 2.800 UF")
      print("|  6 | | | | |    *   Tipo D, 3.500 UF")
      print("|____________|")
      print("|  5 | | | | |")
      print("|____________|")
      print("|  4 | | | | |")
      print("|____________|")
      print("|  3 | | | | |")
      print("|____________|")
      print("|  2 | | | | |")
      print("|____________|")
      print("|  1 | | | | |")
      print("|____________|")

      print("\n---COMPRAR DEPARTAMNETO---")
      rut = int(input("Ingrese rut sin puntos ni guiones y sin numero verificador: "))
      piso = int(input("Ingrese el piso que desea comprar: "))
      tipo = input("Escoja la letra que desea y este diponible: ").capitalize()
      cantidad = int(input("Ingrese la cantidad de departamentos que quiere:"))
      if tipo == "A":
        print("Su total es de:",A + total)
      elif tipo == "B":
        print("Su total es de:",B + total)
      elif tipo == "C":
        print("Su total es de:",C + total)
      elif tipo == "D":
        print("Su total es de:",D + total)
      else:
        print("Letra incorrecta")
        continue
      rut = validar_rut(rut)
      cliente_comprador[rut] = [piso, tipo]
      print(cliente_comprador)
      print("---DEPARTAMENTO COMPRADO---")
      continue

    elif opcion == 2:
      print("---MOSTRAR DEPARTAMENTOS DISPONIBLES---")
      print(" ____________")
      print("|piso| tipo  |")
      print("|    |A|B|C|D|")
      print("| 10 | | | | |")
      print("|____________|")
      print("|  9 | | | | |")
      print("|____________|")
      print("|  8 | | | | |")
      print("|____________|")
      print("|  7 | | | | |")
      print("|____________|")
      print("|  6 | | | | |")
      print("|____________|")
      print("|  5 | | | | |")
      print("|____________|")
      print("|  4 | | | | |")
      print("|____________|")
      print("|  3 | | | | |")
      print("|____________|")
      print("|  2 | | | | |")
      print("|____________|")
      print("|  1 | | | | |")
      print("|____________|")
    elif opcion == 3:
      print("---VER LISTADO DE COMPRADORES---")
      print("estos son los compradores:",cliente_comprador)

    elif opcion == 4:
      print("---MOSTRAR VENTAS TOTALES---")
      print("  _________________________________________")
      print(" | Tipo de departamento | Cantidad | total |")
      print(" | ________________________________________|")
      print(f"| Tipo A 3.800 UF      |{cantidad}|{A}|")
      print(" |_________________________________________|")
      print(f"| Tipo B 3.000 UF      |{cantidad}|{B}|")
      print(" |_________________________________________|")
      print(f"| Tipo C 2.800 UF      |{cantidad}|{C}|")
      print(" |_________________________________________|")
      print(f"| Tipo D 3.500 UF      |{cantidad}|{D}|")
      print(" |_________________________________________|")
      print(f"| Total               |{cantidad}|{total_todo}|")
      print(" |_________________________________________|")
    else:
      print("La opcion ingresada no es valida")
      continue
  import datetime

  hora_actual = datetime.datetime.now()


  print("Saliendo del sistema, Brandon Ramirez ,", hora_actual)

```
